```python
class Student:
    def __init__(self, studentid, sname, m1, m2, m3):
        self.studentid = studentid
        self.sname = sname
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
        self.gpa = self.calculate_gpa()

    def calculate_gpa(self):
        return (1/3 * self.m1) + (1/2 * self.m2) + (1/4 * self.m3)

    def display(self):
        print(f"ID: {self.studentid}, Name: {self.sname}, Marks: ({self.m1}, {self.m2}, {self.m3}), GPA: {self.gpa:.2f}")

def display_all(students):
    if not students:
        print("No students in the list.")
        return
    print("--- Displaying All Students ---")
    for student in students:
        student.display()

def search_by_id(students):
    sid = input("Enter Student ID to search: ")
    found = False
    for student in students:
        if student.studentid == sid:
            print("--- Student Found ---")
            student.display()
            found = True
            break
    if not found:
        print(f"Student with ID '{sid}' not found.")

def search_by_name(students):
    name_search = input("Enter Student Name to search: ").lower()
    found_students = []
    for student in students:
        if name_search in student.sname.lower():
            found_students.append(student)

    if not found_students:
        print(f"No student found with name containing '{name_search}'.")
    else:
        print(f"--- Found {len(found_students)} Student(s) ---")
        for student in found_students:
            student.display()

def calculate_student_gpa(students):
    sid = input("Enter Student ID to view GPA: ")
    found = False
    for student in students:
        if student.studentid == sid:
            print(f"--- GPA for {student.sname} (ID: {student.studentid}) ---")
            print(f"GPA: {student.gpa:.2f}")
            found = True
            break
    if not found:
        print(f"Student with ID '{sid}' not found.")

def main():
    student_list = []
    student_list.append(Student("S101", "Alice", 80, 90, 70))
    student_list.append(Student("S102", "Bob", 75, 85, 95))
    student_list.append(Student("S103", "Charlie", 90, 80, 88))

    while True:
        print("\n--- Student Management System ---")
        print("1. Display all students")
        print("2. Search by ID")
        print("3. Search by name")
        print("4. Calculate the GPA of a student")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            display_all(student_list)
        elif choice == '2':
            search_by_id(student_list)
        elif choice == '3':
            search_by_name(student_list)
        elif choice == '4':
            calculate_student_gpa(student_list)
        elif choice == '5':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```

```python
people = {'Arham': 'Blue', 'Lisa': 'Yellow', 'Vinod': 'Purple', 'Jenny': 'Pink'}
print(f"Original dictionary: {people}")
num_students = len(people)
print(f"1. Number of students: {num_students}")
people['Lisa'] = 'Green'
print(f"2. After changing Lisa's color: {people}")

if 'Jenny' in people:
    del people['Jenny']
    print(f"3. After removing Jenny: {people}")
else:
    print("3. Jenny not found in the dictionary.")

print("4. Students sorted by name:")
sorted_names = sorted(people.keys())
for name in sorted_names:
    print(f"  {name}: {people[name]}")
```
