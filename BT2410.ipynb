bài 1: maxmod

```python
t = int(input())
for _ in range(t):
    n = int(input())
    print(n - (n // 2 + 1))
```

bài 2: inv2x

```python
n = int(input())
a = list(map(int, input().split()))

count = 0
for i in range(n):
    for j in range(n):
        if i<j and a[i] > 2 * a[j]:
            count += 1

print(count)
```

bài 3: pnumber

```python
A, B = map(int, input().split())
for n in range(A, B + 1):
    if n < 2:
        continue

    is_prime = True
    for d in range(2, n):
        if n % d == 0:
            is_prime = False
            break 
    if is_prime:
        print(n)
```
